# 1. Предобработка логов

```python
# vibo: папка где лежат сырые логи
# vibo: ВАЖНО!!! ФАЙЛЫ В ПАПКЕ ДОЛЖНЫ БЫТЬ ОТСОРТИРОВАНЫ ОТ СТАРЫХ К НОВЫМ
# (latest.log...ok - должен прочитаться последним)
link = '/content/drive/MyDrive/minecraft/raw_logs/'
```

```python
# vibo: задаем имена установленных плагинов
plug_names = ['WorldEdit', 'WorldGuard', 'HolographicDisplays']
```

```python
# vibo: лист для имен файлов в папке
file_names = []
# vibo: лист для логов
raw_logs = []

for root, dirs, files in os.walk(link):  
    for filename in files:
      file_names.append(filename)

file_names = sorted(file_names)

for filename in file_names:
  # vibo: в папке файлы двух видов *log.gz и *.log (один последний)
  # vibo: обрабатываем файлы *log.gz
  if filename.endswith("gz"):
    # vibo: создаем путь до файла
    path = link + filename
    # vibo: разархивируем и читаем с учетом кодировки
    with gzip.open(path, 'rb') as f:
      with io.TextIOWrapper(f, encoding='utf-8') as decoder:
        # vibo: в content - полное содержимое одного лога
        content = decoder.read()
        # vibo: обрезаем перенос строк
        for i in content.split('\n'):
          # vibo: выделяем из имени файла дату и добавляем ее в начало строки
          raw_logs.append(filename[:-9] + ' ' + i)
          # vibo: сохраняем крайнюю дату она понадобится ниже для последнего файла без даты
          last_date = filename[:-9]
    print(f'{filename}...ok' )

  # vibo: это обработка файла latest.log
  else:
    with open(link + 'latest.log', encoding="UTF-8") as file_in: 
        for line in file_in:
            messege = line.rstrip("\n")
            raw_logs.append(last_date + ' ' + messege)
    print(f'{filename}...ok' )
```

```python
# vibo: проверяем последнюю строку
raw_logs[3304]
```

```python
# vibo: делаем датафрейм
raw_logs = pd.DataFrame({
    'raw_logs': raw_logs
})
raw_logs
```

```python
# vibo: выделяем дату
raw_logs['date'] = raw_logs['raw_logs'].map(lambda x: x[:11])
# vibo: сохраняем в отдельный столбец оставшиеся данные
raw_logs['raw_logs_split_0'] = raw_logs['raw_logs'].map(lambda x: x[11:])
raw_logs
```

```python
# vibo: проверяем длину каждой строки, где-то остались пустые строки (len = 0)
raw_logs['__len__'] = raw_logs['raw_logs_split_0'].map(lambda x: len(x))
raw_logs
```

```python
raw_logs[raw_logs['__len__'] == 0]
```

```python
# vibo: удалим 23 строки
raw_logs[raw_logs['__len__'] == 0].count()
```

```python
# vibo: удаляем 0, получили 5083 информативных сообщения
raw_logs = raw_logs[raw_logs['__len__'] != 0]
raw_logs
```

```python
# vibo: выделяем время
raw_logs['time'] = raw_logs['raw_logs_split_0'].map(lambda x: x.split('] [')[0][1:])
# vibo: проверяем длину строки в столбце времени
raw_logs['__len__'] = raw_logs['time'].map(lambda x: len(x))
raw_logs
```

```python
# vibo: находим нестандартный лог (Reason: 1с) без времени, 10 значений
raw_logs[raw_logs['__len__'] != 8]
```

```python
# vibo: удалим сообщения без времени
raw_logs = raw_logs[raw_logs['__len__'] == 8]
```

```python
# vibo: итого осталось сообщений
raw_logs
```

```python
# vibo: сохраняем часть лога без времени в отдельный столбце
raw_logs['raw_logs_split_1'] = raw_logs['raw_logs_split_0'].map(lambda x: x.split('] [')[1])
# vibo: удаляем промежуточный столбец
raw_logs = raw_logs.drop(['__len__'], axis=1)

raw_logs
```

```python
# vibo: выделяем стандартные сообщения сервера, остальныю часть сохраняем в отдельный столбец
raw_logs['serv_teg'] = raw_logs['raw_logs_split_1'].map(lambda x: x.split(']:')[0])
raw_logs['raw_logs_split_2'] = raw_logs['raw_logs_split_1'].map(lambda x: x.split(']:')[1])

raw_logs
```

```python
# vibo: заполняем имена плагинов для сообщений
raw_logs['plug_name'] = raw_logs['raw_logs_split_2'].map(lambda x: [y for y in plug_names if y in x])
# vibo: делаем вспомогательный столбец, если плагин есть 1, нет 0
raw_logs['__plug__'] = raw_logs['plug_name'].map(lambda x: 1 if len(x) == 1 else 0)

raw_logs
```

```python
# vibo: заменил [] на NaN
raw_logs['plug_name'] = raw_logs[raw_logs['__plug__'] == 1]['plug_name'].map(lambda x: x[0])

raw_logs
```

```python
# vibo: основные логи сервера в отдельный столбец
raw_logs['serv_mess'] = raw_logs[raw_logs['__plug__'] == 0]['raw_logs_split_2']

raw_logs
```

```python
# vibo: логи плагинов сложил в отдельный столбец
raw_logs['plug_mess'] = raw_logs[raw_logs['__plug__'] == 1]['raw_logs_split_2'].map(lambda x: x[(x.find(']')) + 2:])

raw_logs[['date', 'time', 'serv_teg', 'serv_mess', 'plug_name', 'plug_mess']]
```
